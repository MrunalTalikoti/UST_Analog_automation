```python
import pandas as pd

gain_df = pd.read_csv(r"E:\VIIT\TY\UST Analog Workflow Automation\Project\UST_Analog_automation\data\processed\Opam\gain_ml_ready.csv")
ugf_df = pd.read_csv(r"E:\VIIT\TY\UST Analog Workflow Automation\Project\UST_Analog_automation\data\processed\Opam\ugf_ml_ready.csv")
pm_df = pd.read_csv(r"E:\VIIT\TY\UST Analog Workflow Automation\Project\UST_Analog_automation\data\processed\Opam\pm_ml_ready.csv")


merged_df = (
    gain_df
    .merge(ugf_df, on=["W", "X", "Y"], how="inner")
    .merge(pm_df,  on=["W", "X", "Y"], how="inner")
)

output_path = r"E:\VIIT\TY\UST Analog Workflow Automation\Project\UST_Analog_automation\data\processed\Opam\merged_ml_ready.csv"

merged_df.to_csv(output_path, index=False)

print("✅ Saved merged dataset:", output_path)
print("Rows:", len(merged_df))
```
