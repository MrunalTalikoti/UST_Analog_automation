```python
import pandas as pd
```

```python
df  = pd.read_csv(r'D:\UST Project\UST_Analog_automation\data\raw\CSAmp.csv')
```

```python
df = df.rename(columns={df.columns[0]: "PMOS_W"})
```

```python
df_long = df.melt(
    id_vars="PMOS_W",
    var_name="NMOS_W",
    value_name="Gain"
)
```

```python
df_long["PMOS_W"] = pd.to_numeric(df_long["PMOS_W"])
df_long["NMOS_W"] = pd.to_numeric(df_long["NMOS_W"])
df_long["Gain"]   = pd.to_numeric(df_long["Gain"])
```
